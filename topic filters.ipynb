```python
scope = ['https://www.googleapis.com/auth/webmasters.readonly']
credentials = service_account.Credentials.from_service_account_info(json_key, scopes=scope)
service = build(
    'webmasters',
    'v3',
    credentials=credentials
)
```

```python
# Read keywords from the uploaded Excel file
df = pd.read_excel('topic filters 1.xlsx')
keywords = df['keywords'].tolist()

# Create lists to store metrics
keyword_list = []
clicks_list = []
impressions_list = []
ctr_list = []

# Loop through keywords and retrieve performance data
for keyword in keywords:
    request = {
        'startDate': '2023-04-01',
        'endDate': '2023-06-30',
        'dimensions': ['query'],
        'dimensionFilterGroups': [{
            'filters': [{'dimension': 'query', 'operator': 'equals', 'expression': keyword}]
        }]
    }
    response = service.searchanalytics().query(siteUrl='https://www.bajajfinserv.in/', body=request).execute()

    # Extract and analyze the performance metrics from the response
    rows = response.get('rows', [])

    if rows:
        clicks = rows[0].get('clicks', 0)
        impressions = rows[0].get('impressions', 0)
        click_through_rate = (clicks / impressions) * 100
    else:
        clicks = 0
        impressions = 0
        click_through_rate = 0.0

    # Append metrics to lists
    keyword_list.append(keyword)
    clicks_list.append(clicks)
    impressions_list.append(impressions)
    ctr_list.append(click_through_rate)

# Create a DataFrame from the lists
data = {
    'Keyword': keyword_list,
    'Clicks': clicks_list,
    'Impressions': impressions_list,
    'CTR': ctr_list
}
result_df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
result_df.to_excel('search_console_metrics.xlsx', index=False)
```
